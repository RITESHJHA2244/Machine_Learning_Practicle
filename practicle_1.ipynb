```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load data
iris = load_iris()
x = iris.data[:50, 0]   # Sepal Length
y = iris.data[:50, 1]   # Sepal Width

# Reshape x
x = x.reshape(-1, 1)

# Design matrix X = [x, 1]
X = np.hstack((x, np.ones((x.shape[0], 1))))

# Convert y to column vector
Y = y.reshape(-1, 1)

# Normal Equation
w = np.linalg.inv(X.T @ X) @ X.T @ Y

# Extract slope and intercept
m = w[0, 0]
c = w[1, 0]

print("Weight vector w:")
print(w)
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
m = 0.45   # slope
c = 1.25   # intercept

# x values
x = np.linspace(4, 7, 100)

# equation y = mx + c
y = m * x + c

# Plot line
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line Plot of y = mx + c")
plt.grid(True)
plt.show()
```
